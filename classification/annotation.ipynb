```python
import pandas as pd
train_annotations=pd.read_json('train_annotations.json')
train_annotations.head()
```

```python
test_annotations=pd.read_json('valid_annotations.json')
test_annotations.head()
```

```python
train_annotations=train_annotations[['image_id','category_id']]
train_annotations.head()
```

```python
test_annotations=test_annotations[['image_id','category_id']]
test_annotations.head()
```

```python
train_annotations['category_id']=train_annotations['category_id'].replace({1:'Penguin',2:'Turtle'})
train_annotations.head()
```

```python
test_annotations['category_id']=test_annotations['category_id'].replace({1:'Penguin',2:'Turtle'})
test_annotations.head()
```

```python
Id=[]
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('train'):
    for filename in filenames:
        Id.append(os.path.join(dirname, filename))
Id[:5]
```

```python
train=pd.DataFrame()
train=train.assign(filename=Id)
train.head()
```

```python
train['image_id']=train['filename']
train['image_id']=train['image_id'].str.replace('train\image_id_','')
train['image_id']=train['image_id'].str.replace('.jpg','')
train['image_id']=train['image_id'].astype(int)
train.head()
```

```python
labels = train.sort_values('label')
labels.head()
```

```python
class_names = list(labels.label.unique())
class_names[:5]
```

```python
newpath = r'./train' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
import shutil, os
#creating subfolders
for c in class_names:
    dest =  r'./tmp/train/'+str(c)
    os.makedirs(dest)
    for i in list(labels[labels['label']==c]['filename']): # Image Id
        get_image = os.path.join('train', i) # Path to Images 
        move_image_to_cat = shutil.copy(get_image, dest)
```
